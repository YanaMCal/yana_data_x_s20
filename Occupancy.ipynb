```python
import pandas as pd
import numpy as np
```

```python
occ = pd.read_csv('Building Occupancy EDITED.csv')
occ.head(10)
```

```python
blds = pd.read_csv('BLD_FEATURES.csv')
blds.head(20)
```

```python
blds["Building_name"][17]="Le Conte Hall"
```

```python
occ.head(30)
```

```python
def split_str(string):
    return string.split('-')[0]

bld_occ = occ.copy()
bld_occ["Bld_name"]=bld_occ["Desc"].apply(split_str)
bld_occ.head(10)
```

```python
uniq=[]
group = 0
for s in blds["Building_name"]:
    for i in bld_occ["Bld_name"]:
        if i in s:
            if s not in uniq:
                uniq.append(s)
```

```python
uniq
```

```python
bld_occ=bld_occ[bld_occ['Bld_name'].isin(uniq)]
```

```python
bld_occ
```

```python
groupped = bld_occ[["Desc","COUNT(Employee ID)","Bld_name"]].groupby(["Bld_name","Desc"]).agg(max)
```

```python
groupped = groupped.groupby(level=[0]).sum()
```

```python
groupped.to_csv('occ_cleaned.csv')
```
