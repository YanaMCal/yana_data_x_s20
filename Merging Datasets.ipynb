## Separating data sets
We define the instructional semester date range, then split the data into various sets: instructional weekdays, academic holidays, breaks, etc. We will continue the data cleaning to exclude extreme outliers within these sets.

So far, we exclude academic holidays and the PGE power outage on UCB campus in our semester date ranges.

```python
# Instruction ranges 
sp_start = datetime.strptime('1/22/2019', '%m/%d/%Y')
sp_end = datetime.strptime('5/17/2019', '%m/%d/%Y')
sp_range = [sp_start, sp_end]

su_start = datetime.strptime('05/28/2019', '%m/%d/%Y')
su_end = datetime.strptime('08/16/2019', '%m/%d/%Y')
su_range = [su_start, su_end]

fa_start = datetime.strptime('8/28/2019', '%m/%d/%Y')
fa_end = datetime.strptime('12/20/2019', '%m/%d/%Y')
fa_range = [fa_start, fa_end]

# Holidays
sp_holidays = [
    '02/18/2019',
    '03/25/2019',
    '03/25/2019',
    '03/26/2019',
    '03/27/2019',
    '03/28/2019',
    '03/29/2019'
]

su_holidays = ['07/04/2019']

# fall holidays including PGE power outage on UCB campus
fa_holidays = [
    '09/02/2019', 
    '10/26/2019',
    '10/27/2019',
    '10/28/2019',
    '11/11/2019', 
    '11/27/2019', 
    '11/28/2019', 
    '11/29/2019'
]

# combing the holidays for 2019 including the spring, summer, fall semesters
holidays = sp_holidays + su_holidays + fa_holidays


# convert strings into datetime
holidays_dt = [datetime.strptime(x, '%m/%d/%Y') for x in holidays]

# check if date in summer dates
sp_range =  pd.date_range(start=sp_range[0], end=sp_range[1])
su_range =  pd.date_range(start=su_range[0], end=su_range[1])
fa_range =  pd.date_range(start=fa_range[0], end=fa_range[1])

all_range = sp_range.append(su_range).append(fa_range)
```

```python
#exclude all holidays and outages from the instructional dates range
clean_dates = all_range[~all_range.isin(holidays)]
clean_dates
```

```python
def clean_data(df):
    df['Day'] = df['Date_Time'].str.split()
    df['Day'] = df['Day'].apply(lambda x: x[0])
    
    #convert Day into datetime object to select only eligigble dates
    df["Day"] = pd.to_datetime(df["Day"])
    df = df[df["Day"].isin(clean_dates)]
    df['Date_Time'] = pd.to_datetime(df['Date_Time'])

    #Converting datetime into boolean variable to represent weekdays vs weekends
    #0 : Monday, ..., 6 : Sunday
    #weekday = 1, if the day of the week is a weekday; weekday = 0, o.w.
    #sp = spring, su = summer, fa = fall, wi = winter
    dates = []
    weekday = []
    seasons = []
    for index in range(len(df['Date_Time'])):
        date = df['Date_Time'].iloc[index]
        dates += [date]
        if date.weekday() > 4:
            weekday += [0]
        else:
            weekday += [1]
        if sp_start <= date <= sp_end:              
            seasons += ['sp']
        elif su_start <= date <= su_end:
            seasons += ['su']
        elif fa_start <= date <= fa_end:            
            seasons += ['fa']
        else:
            seasons += ['wi']
    df['weekday'] = weekday
    df['season'] = seasons
    
    
    return df
```
